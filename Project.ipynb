```python
# A simple password generator for office files
import random

pword=str(input("input your name to generate password"))


if len(pword)>=2:
    print(f"{pword}{random.randint(1000,10000)}")
else:
    print("name characters not up to requirement")


    
```

```python
#creating a password for email/ office use to test security and strength

pw = str(input("write your password"))


if len(pw)<=5:
    print("Invalid: Must contain at least one digit")
if not any(char.isdigit() for char in pw):
    print("add number to your password")   
if not any(char.islower() for char in pw):        
    print("add lowercase character to your password")
if not any(char.isupper() for char in pw):
    print("add upper case character to your password")
if not any(char in '!@#$%^&*' for char in pw):    
    print("add special characters")
print("the password", pw,"is strong")
        

  
    
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sales=pd.read_csv('supermarket_sales.csv') 
sales.tail(6)

```

```python

#the mean of the last five index
total_mean=sales['Total'].mean()
print(f"The result of the total mean is {total_mean:.2f}")

#the median of the last five index
total_median=sales['Total'].median()
print(f"The result of the total median is {total_median:.2f}")

#the standard deviation of the last five index
total_std=sales['Total'].std()
print(f"The result of the total std is {total_std:.2f}")

#the variances of the last five index
total_vr=sales['Total'].var() 
print(f"The result of the total std is {total_vr:.2f}")
```
